## A. Distribution of Calories
The histogram and boxplot for the said distribution can be plotted together with the help of a subplot, so that we can compare both the visuals side by side to get more comprehensive insights about the distribution as well as the outliers.

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(mcd['Calories'], bins=10, kde=True)
plt.title('Histogram of Calorie Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=mcd['Calories'])
plt.title('Box Plot of Calorie Distribution')
plt.xlabel('Calories')

plt.tight_layout()
plt.show()
```

The shape of the histogram is right-skewed, with a peak at approximately 125-500 calories and a long tail towards the higher values of calories. The majority of menu items fall on the lower end of the calorie count, and there are only a few high-calorie outliers. The observations are confirmed by the box plot, where the median is approximately 340 calories and the 25th and 75th percentiles are 210 and 500, respectively. The box plot reveals the presence of several outliers, some of which are associated with high calories, starting from the "McFlurry with Reese's Peanut Butter Cups (Medium)" at 810 calories.

# 4.2 Compare nutritional characteristics of different food categories (e.g., burgers, salads, desserts).

### Nutritional Navigator: Stacked Bar Chart and Table Revealing Healthy vs. Unhealthy Food Categories

```python
plt.figure(figsize=(12, 8))
mcd.groupby('Category')[['Calories', 'Carbohydrates', 'Dietary Fiber', 'Protein',
                    'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                    'Calcium (% Daily Value)','Sodium', 'Iron (% Daily Value)']].median().plot(kind='bar',stacked=True,legend=False)
plt.ylabel('Nutritional values')
plt.title('Calories and nutritional content across food Categories')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='center left', bbox_to_anchor=(1.04, 0.5), 
           title='State', fontsize='small')
plt.savefig("healthy.jpg", bbox_inches='tight') 
plt.show()
mcd.groupby('Category')[['Calories', 'Carbohydrates', 'Dietary Fiber', 'Protein',
                    'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                    'Calcium (% Daily Value)','Sodium', 'Iron (% Daily Value)']].median()
```

```python
plt.figure(figsize=(12, 8))
mcd.groupby('Category')[['Total Fat','Cholesterol','Sugars']].median().plot(kind='bar',stacked=True,legend=False)
plt.ylabel('content')
plt.title('Total Fat,Sugar,Chlesterol level across Food Categories')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='center left', bbox_to_anchor=(1.04, 0.5), 
           title='State', fontsize='small')
plt.savefig("unhealthy.jpg", bbox_inches='tight')
plt.show()
mcd.groupby('Category')[['Total Fat','Cholesterol','Sugars']].median()
```

The stacked bar charts and tables show the average nutrient content for each food category. They help identify healthier options and point out foods high in fat, cholesterol, and sugars that should be limited.

### 5.2 Determine the average nutritional content of popular menu categories.

```python
nutritional_values = [
    'Calories', 'Carbohydrates', 'Dietary Fiber', 'Protein', 
    'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 
    'Calcium (% Daily Value)', 'Sodium', 'Iron (% Daily Value)'
]

category_means = mcd.groupby('Category')[nutritional_values].mean()

category_totals = category_means.sum(axis=1)

plt.figure(figsize=(6, 6))
plt.pie(category_totals, labels=category_totals.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Total Average Nutrition by Food Category')
plt.axis('equal')
plt.savefig("pie chart.png")
plt.show()
```
